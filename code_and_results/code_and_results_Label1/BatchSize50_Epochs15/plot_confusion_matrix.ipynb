```python
%reset -f

from sklearn.metrics import confusion_matrix
import pandas as pd

all_data = pd.read_csv('confusion_matrix1.csv', delimiter = ";", encoding = 'latin_1')

y_test=all_data['Label1']
y_pred=all_data['Label1_pred']

#Get the confusion matrix
cf_matrix = confusion_matrix(y_test, y_pred)

print(cf_matrix)
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

#ax = sns.heatmap(cf_matrix, annot=True, cmap='Blues')

ax = sns.heatmap(cf_matrix/np.sum(cf_matrix), annot=True, 
            fmt='.2%', cmap='Blues')

ax.set_title('Confusion Matrix Label 1 \n\n');
ax.set_xlabel('\nPredicted Category')
ax.set_ylabel('Actual Category ');

## Ticket labels - List must be in alphabetical order
ax.xaxis.set_ticklabels(['Not classifiable','Neutral', 'Positive', 'Negative'])
ax.yaxis.set_ticklabels(['Not classifiable','Neutral', 'Positive', 'Negative'])

## Display the visualization of the Confusion Matrix.
plt.show()

fig = ax.get_figure()
fig.savefig("plot1.pdf")
```
